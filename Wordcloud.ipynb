```python
import pandas as pd
import numpy as np
import os # provides functions for interacting with the operating system
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
text = open("/content/cv.txt").read()
print(text)
stopwords = set(STOPWORDS)
stopwords.update(["Torrado", "Romero", "Profile", "Duarte", "Cecile", "+351936004117", "+41719441912", "de"])
```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords).generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(stopwords=stopwords, max_font_size=70, background_color="white", width=600, height=400,).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file("wordcloud_PS.png")
```

```python
from PIL import Image
plant_mask = np.array(Image.open("plant.jpg"))
plant_mask
#plant_mask[230:250, 240:250]
plt.imshow(plant_mask)
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white",
                  max_words=1000,
                  mask=plant_mask,
                  stopwords=stopwords,
                  contour_width=2,
                  repeat=True,
                  max_font_size=50,
                  contour_color='grey')

# Generate a wordcloud
wc.generate(text)

# store to file
wc.to_file("plant.jpg")

# show
plt.figure(figsize=[10,5])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
